```python
from keras.models import load_model
model = load_model('zero_model.h5')
```

```python
import pandas as pd
import numpy as np
import keras
```

```python
df = pd.read_csv('data/review_dataset.csv')
df.head()
df.tail()
```

```python
from keras.preprocessing.text import text_to_word_sequence
# define the document
text = 'The quick brown fox jumped over the lazy dog.'
# tokenize the document
result = text_to_word_sequence(df['review'][0])
print(result)
```

```python
df['review'] = df['review'].astype('str')
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

t = Tokenizer()
# fit the tokenizer on the documents
t.fit_on_texts(df['review'])
```

```python
example = t.texts_to_sequences(df['review'])
example = pad_sequences(example, maxlen=128)

incorrects = []
pred = model.predict(example)

print type(pred)
```

```python
header = ['FB', 'FO', 'HK', 'HOTEL']
df2 = pd.DataFrame(pred)
```

```python
df2.columns = header
df2.head()
```

```python
df['FB'] = df2['FB']
df['FO'] = df2['FO']
df['HK'] = df2['HK']
df['HO'] = df2['HOTEL']

df.tail()
```
